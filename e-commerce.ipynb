```python
#4. Сколько у каждого из пользователей в среднем покупок в неделю (по месяцам)? 
# Не стоит забывать, что внутри месяца может быть не целое количество недель. 
# Например, в ноябре 2021 года 4,28 недели. И внутри метрики это нужно учесть.
```

```python
cus_orders.order_purchase_timestamp = pd.to_datetime (cus_orders.order_purchase_timestamp) 
#переводим колонку (order_purchase_timestamp —  время создания заказа) в тип даты
```

```python
cus_orders['weeks'] = cus_orders.order_purchase_timestamp.dt.days_in_month / 7 #отбираем из этой колонки количество недель в месяце
```

```python
cus_orders['month'] = cus_orders.order_purchase_timestamp.dt.to_period("M") #отбираем из этой колонки только месяц
```

```python
cus_orders.head()
```

```python
#посчитаем количество покупок у каждого уникального пользователя в месяц
m_orders_count = cus_orders.query('order_status == "delivered"') \
                         .groupby(['customer_unique_id', 'month', 'weeks'], as_index = False) \
                         .agg({'order_id' :'count'}) \
                         .sort_values('order_id', ascending = False)
```

```python
m_orders_count['order_id'] = m_orders_count.order_id / m_orders_count.weeks
```

```python
#Ответ в колонке 'order_id'
m_orders_count
```

```python
#5. Используя pandas, проведи когортный анализ пользователей. 
# В период с января по декабрь выяви когорту с самым высоким retention на 3й месяц.
```

```python
#получаем уникальные связки "Клиент Год/Месяц" с покупками
df1 = cus_orders.query('order_status == "delivered"')
```

```python
df1 = cus_orders[['customer_unique_id', 'month', 'order_id']]
```

```python
df1.head()
```

```python
#для каждого клиента посчитаем Год/Месяц первой покупки
df2 = cus_orders.query('order_status == "delivered"') \
          .groupby(['customer_unique_id', 'month'], as_index = False) \
          .agg({'month' : 'min'}) \
          .rename(columns = {'month' : 'first_order_month'})
```

```python
df2.head()
```

```python
#смёрджим оба датафрейма по номеру клиента
joint = df1.merge(df2, on = 'customer_unique_id')
```

```python
joint
```

```python
#в разрезе "Год/месяц ПЕРВОЙ покупки Год/Месяц покупки" посчитаем количество уникальных клиентов
joint1 = joint.groupby(['first_order_month', 'month'], as_index = False).agg({'customer_unique_id' : 'nunique'}) 
```

```python
joint1.head()
```

```python
from operator import attrgetter
```

```python
joint1['period'] = (joint1.month - joint1.first_order_month).apply(attrgetter("n"))
#считаем месяц покупки, начинаем с нулевого - месяц первой покупки
```

```python
joint1.sort_values('period')
```

```python
joint1 = joint1.dropna(axis='index', how='any', subset=['month'])
#Удаляем пропуски
```
